  1. Write a program that creates gramatically valid English sentences.
    * Here, we consider a sentence to be gramatically correct when it follows the simple English grammar of the form: `Article Adjective Noun Verb.` That is, even the sentence `A insect fly.` is, for the moment, considered correct. Use the files inside the data folder: nouns, verbs and adjectives. HINT: read them at put content into 3 lists.
  2. Extend the above program to generate all possible sentences with the given words.

```python
# Importing all verbs

with open('../data/verbs') as verbs_file:
    verbs = verbs_file.read().split('\n')
# print(verbs)
```

```python
# Importing all nouns

with open('../data/nouns') as nouns_file:
    nouns = nouns_file.read().split('\n')
# print(nouns)
```

```python
# Importing all adjectives

with open('../data/adjectives') as adjectives_file:
    adjectives = adjectives_file.read().split('\n')
# print(adjectives)
```

```python
import random

def get_sentence():
    adj = adjectives[random.randint(0,100)]
    noun = nouns[random.randint(0,100)]
    verb = verbs[random.randint(0,100)]
    return f'{adj.capitalize()} {noun} {verb}'


print(get_sentence())
```

```python
def all_sentences_adj(adj):
    return [f'{adj.capitalize()} {noun} {verb}' for noun in nouns for verb in verbs]
print(all_sentences_adj(adjectives[55]))
    
```

```python
def all_sentences_noun(noun):
    return [f'{adj.capitalize()} {noun} {verb}' for adj in adjectives for verb in verbs]
print(all_sentences_noun(nouns[33]))
```

```python
def all_sentences_verb(verb):
    return [f'{adj.capitalize()} {noun} {verb}' for adj in adjectives for noun in nouns]
print(all_sentences_verb(verbs[22]))
```
