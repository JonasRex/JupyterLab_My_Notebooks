## Find similarities
I wanna make a training model that I can train with my data. I will make a bunch of list with the following names:
- interesting_pairs # Example: 'Beyoncé' and 'Beyoncé Knowles'. I want this count as the same person. 
- approved_list # When ever i manually approve a match, i want the program to remember, so I don't have to do it again.
- ignoring_list # Common names like Max, might pop up alot. So this list automatically remove them from future checks

It should result in a list that I can apply the proper changes to my dataset. So the singer / writers get the real credit.

```python
other_connections = [['Lennon','Lennon-McCartney','John Lennon'],['McCartney','Lennon-McCartney','John McCartney']]

interesting_pairs = []

for x in all_people:
    for y in all_people:
        if(x in y and x != y):
            interesting_pairs.append([x,y])
            #print(x, " is similar to ", y)

len(interesting_pairs)
```

```python
# OBS: dont run this, unless you want to reset the lists!
ignoring_list = [["Train","Meghan Trainor"]]
approved_list = [['Beyoncé', 'Beyoncé Knowles']]
list_to_change = []
```

```python
print('Interesting pairs before first sorting: ',len(interesting_pairs))

for x in interesting_pairs:
    if(x in ignoring_list):
        interesting_pairs.remove(x)
    elif(x in approved_list):
        interesting_pairs.remove(x)
        list_to_change.append(x)

print('Interesting pairs affter first sorting: ',len(interesting_pairs))
```

```python
print("Interesting pairs:",len(interesting_pairs))
print("Approved:",len(approved_list))
print("Ignored:",len(ignoring_list))
print("List ready for change:",len(list_to_change))
```

```python
for [x,y] in interesting_pairs:
    answer = input(f'Is: {x} the same as: {y}?')
    if(answer.lower() == 'y'):
        approved_list.append([x,y])
    elif(answer.lower() == 'n'):
        ignoring_list.append([x,y])
    elif(answer.lower() == 'm'):
        print('Maybe. Not implemented yet')
    else:
        print('Wrong input')
```

```python
print("Interesting pairs:",len(interesting_pairs))
print("Approved:",len(approved_list))
print("Ignored:",len(ignoring_list))
print("List ready for change:",len(list_to_change))
```

```python
print('Interesting pairs before second sorting: ',len(interesting_pairs))

for x in interesting_pairs:
    if(x in ignoring_list):
        interesting_pairs.remove(x)
    elif(x in approved_list):
        interesting_pairs.remove(x)
        list_to_change.append(x)

print('Interesting pairs after second sorting: ',len(interesting_pairs))
```

```python
def convert_artist_data_into_df(artist):
    data_list = []
    for name in all_people:
        main_artist = len(df[df['Artist'] == name])
        featuring_artist = len(df[df['Featuring'] == name])
        writer = len(df[df['Writer 1'] == name]) + len(df[df['Writer 2'] == name])
        data_list.append([name, main_artist, featuring_artist, writer])
    return pd.DataFrame(data_list, columns= ['Name', 'Main Artist' ,'Featuring Artist', 'Writer'] )


df_to_csv = convert_artist_data_into_df(all_people).sort_values("Name", ascending=True).reset_index(drop=True)
#df_to_csv = convert_artist_data_into_df(all_people).sort_values("Main Artist", ascending=False).reset_index(drop=True)

df_to_csv
```
