# Save all the songs as text files

```python
# Make a folder for each year
import os

try:
    for year in range(1989,1992):
        os.mkdir('data/lyrics/'+str(year))
except Exception as e:
    print(e)
```

```python
def fetch_all_lyrics_from_df(df):
    df['Lyrics'] = df.apply(lambda row: fetch_lyric(row), axis=1)
    
    for index, row in df.iterrows():
        # Only try to save, if not null
        if(row['Lyrics'] != ""):
            save_lyrics(row['Title'], row['Artist'], row['Year'], row['Lyrics'])
    
    return df

def fetch_lyric(row):  
    title = row['Title']
    artist = row['Artist']
    
    return get_lyrics_via_search(title, artist)

def save_lyrics(title, artist, year, lyrics):
    name = get_storing_name(title, artist)
    try:
        with open(f'data/lyrics/{year}/{name}.txt', 'w') as f:
            f.write(lyrics)
    except FileNotFoundError:
        print("The 'docs' directory does not exist")
        
def underscore_title_artist(title, artist):
    return f'{title} {artist}'.replace(" ", "_")

def get_storing_name(title, artist):
    title = check_title(title) # Some title contains '/'
    return underscore_title_artist(title, artist)
```

```python
def get_year(df, year): 
    df_lyrics = df[df['Year'] == year].reset_index(drop=True)
    df_lyrics = df_lyrics[['Year', 'Place', 'Title', 'Artist', 'Genre 1']]
    
    
    result = fetch_all_lyrics_from_df(df_lyrics)
    print("Songs without or too short lyrics:", result[result['Lyrics'] == ''].shape[0])
    
    return result
```

```python
import os

def load_text_files():
    result = []
    
    for year in range(1992,2020):
        files = os.listdir(f'data/lyrics/{year}')
        for file in files:
            
            song = [year, ""]
            with open (f'data/lyrics/{year}/{file}', "r") as myfile:
                data = myfile.read().splitlines()
                for line in data:
                    song[1] += line + " "
            result.append(song)
            
    return result


test = load_text_files()
test
```

```python
# Python3 program Split camel case
# string to individual strings
 
def camel_case_split(str):
    words = [[str[0]]]
 
    for c in str[1:]:
        if words[-1][-1].islower() and c.isupper():
            words.append(list(c))
        else:
            words[-1].append(c)
 
    return [''.join(word) for word in words]
     
# Driver code
str = "GeeksForGeeks"
print(camel_case_split(str))
```

## Practicing Dataframes and Apply method

```python
import pandas as pd

df1 = pd.DataFrame(
    [
        (1, 'Hello', 158, True, 12.8),
        (2, 'Hey', 567, False, 74.2),
        (3, 'Hi', 123, False, 1.1),
        (4, 'Howdy', 578, True, 45.8),
        (5, 'Hello', 418, True, 21.1),
        (6, 'Hi', 98, False, 98.1),
    ],
    columns=['colA', 'colB', 'colC', 'colD', 'colE']
)
print(df1)
```

```python
def categorise(row):  
    if row['colC'] > 0 and row['colC'] <= 99:
        return 'A'
    elif row['colC'] > 100 and row['colC'] <= 199:
        return 'B'
    elif row['colC'] > 200  and row['colC'] <= 299:
        return 'C'
    return 'D'
```

```python
df1['colF'] = df1.apply(lambda row: categorise(row), axis=1)
```
