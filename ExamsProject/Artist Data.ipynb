# Make Artist Data for Part 2 

```python
import pandas as pd
df = pd.read_csv('data/year2009_test.csv')
#df = pd.read_csv('data/year1962_test.csv')

df
```

```python
artist = df['Artist'].dropna().tolist()
#artist = []
featuring = df['Featuring'].dropna().tolist()
#featuring = []
writer1 = df['Writer 1'].dropna().tolist()
#writer1 = []
writer2 = df['Writer 2'].dropna().tolist()
#writer2 = []
all_people = [*artist, *featuring, *writer1, *writer2 ]
all_people = sorted(set(all_people))
all_people[:10]
```

```python
# Check for similarities. Won't find result like Beatles and John Lennon 
for x in all_people:
    for y in all_people:
        if(x in y and x != y):
            print(x, " is similar to ", y)
```

```python
def convert_artist_data_into_df(artist):
    data_list = []
    for name in all_people:
        main_artist = len(df[df['Artist'] == name])
        featuring_artist = len(df[df['Featuring'] == name])
        writer = len(df[df['Writer 1'] == name]) + len(df[df['Writer 2'] == name])
        data_list.append([name, main_artist, featuring_artist, writer])
    return pd.DataFrame(data_list, columns= ['Name', 'Main Artist' ,'Featuring Artist', 'Writer'] ).sort_values('Name', ascending=True)


convert_artist_data_into_df(all_people)
```
