### Class exercise: masking
- For the dataset: `data = np.arange(1,101).reshape(10,10)`

```python
import numpy as np

data = np.arange(1,101).reshape(10,10)
print(data)
```

### 1)

```python
import numpy as np
filename = '../data/befkbhalderstatkode.csv'

bef_stats_df = np.genfromtxt(filename, delimiter=',', dtype=np.uint, skip_header=1)
dd = bef_stats_df

dd
```

### 2)


```python
year = 2015
age = 0
cc = 5180

german_mask = (dd[:,0] == year) & (dd[:,2] == age) & (dd[:,3] == cc)  # Year / Age / Country Code

control = dd[german_mask]
print(f'Number of germans age {age} in KBH in {year}: ',control[:,4].sum())
```

### 3, 4, 5)


```python
def get_population_data(year, area= None, age=None, country=None):
    mask_year = dd[:,0] == year
    mask_area = dd[:,1] == area if area else True
    mask_age = dd[:,2] == age if age else True
    mask_country = dd[:,3] == country if country else True
    
    data = dd[mask_year & mask_area & mask_age & mask_country]
    
    return data[:,4].sum()

    

print('Population (Valby) not age specific in 2014: \t', get_population_data(2014, area=5, country=5100))
print('Population (Valby) of 30 year olds in 2014: \t', get_population_data(2014, 5, 30, 5100))
```

### 6) create a new function that can also give average values for each year if year whas not provided

```python
def get_average_pop(year =None):
    accu_age = 0
    
    
    mask_year = dd[:,0] == year if year else True
    
    
    data = dd[mask_year]
    
    for line in data:
        accu_age += line[2].sum() * line[4].sum()
        
    return accu_age / data[:,4].sum()
    
    #return np.average(data[:,4])
    #return accu_age


print("Average Age in 2015: ", get_average_pop(2015))
```
