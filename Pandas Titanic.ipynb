### 2B use df.apply() to Change all names to the "firstname lastname" format

```python
def change_name(name):
    if ',' in name:
        name = name.split(", ")
        if '.' in name[1]:
            first_name = name[1].split('. ')[1]
        else:
            first_name = name[1]
        return f'{first_name} {name[0]}'
    else:
        return name


    
df['Name'] = df.apply(lambda name: change_name(name['Name']), axis=1)


df.head()
```

### 3A Get only last name by the use of str method split() on comma.

```python
def only_last_name(name):
    if ',' in name:
        name = name.split(", ")
        return name[0]
    else:
        return name
    
df['Name'] = df.apply(lambda name: only_last_name(name['Name']), axis=1)


df.head()    
```
